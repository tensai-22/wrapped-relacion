```python
import re
from collections import Counter
from datetime import datetime
import matplotlib.pyplot as plt

# Ruta del archivo de frases románticas
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"

# Leer el archivo de frases
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Lista ampliada de palabras clave románticas
palabras_romanticas = [
    "amor", "cariño", "te amo", "te quiero", "mi vida", "mi cielo",
    "mi amor", "te extraño", "corazón", "beso", "abrazar", "linda",
    "preciosa", "hermosa", "ñu", "añuñin", "gordita", "te adoro", 
    "princesa", "mi reina", "mi rey", "mi cielo", "mi bebé", "mi niña", 
    "mi niño", "te necesito", "mi todo", "eres especial", "qué linda eres"
]

# Procesar frases
frecuencias = Counter()
mensajes_por_mes = Counter()

for frase in frases:
    # Extraer fecha
    match = re.match(r"\[(\d{2}/\d{2}/\d{2})", frase)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y")
        mes_año = fecha.strftime("%Y-%m")
        mensajes_por_mes[mes_año] += 1

    # Contar palabras clave en la frase
    for palabra in palabras_romanticas:
        if palabra in frase.lower():
            frecuencias[palabra] += 1

# Mostrar estadísticas principales
print("\nEstadísticas de frases románticas:")
print(f"Total de frases románticas: {len(frases)}")
print(f"Palabras románticas más usadas:\n{frecuencias.most_common(5)}")

# Visualización: Frases por mes
plt.figure(figsize=(10, 6))
plt.bar(mensajes_por_mes.keys(), mensajes_por_mes.values())
plt.title("Cantidad de frases románticas por mes")
plt.xlabel("Mes y Año")
plt.ylabel("Cantidad de frases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualización: Palabras más frecuentes
palabras, cantidades = zip(*frecuencias.most_common(10))
plt.figure(figsize=(10, 6))
plt.bar(palabras, cantidades)
plt.title("Palabras románticas más usadas")
plt.xlabel("Palabra")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import re
from collections import Counter
from datetime import datetime
import matplotlib.pyplot as plt

# Rutas de los archivos
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"
ruta_chat = r"C:\Users\user\Desktop\ñucita_cabezona\_chat.txt"

### Análisis 1: Frases románticas ###
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Contadores de "m" y "Ojiñawicha"
contador_romantico = Counter()
caracteres_romanticos = Counter()

for frase in frases:
    match = re.match(r"\[\d{2}/\d{2}/\d{2} \d{2}:\d{2}:\d{2}\] (\w+):", frase)
    if match:
        autor = match.group(1)
        contador_romantico[autor] += 1
        caracteres_romanticos[autor] += len(frase)

# Mostrar estadísticas del análisis 1
print("\nEstadísticas de frases románticas:")
print(f"Mensajes románticos por persona: {contador_romantico}")
print(f"Total de caracteres románticos por persona: {caracteres_romanticos}")

# Visualización: Frases románticas por persona
plt.figure(figsize=(8, 6))
plt.bar(contador_romantico.keys(), contador_romantico.values())
plt.title("Frases románticas por persona")
plt.xlabel("Persona")
plt.ylabel("Cantidad de frases")
plt.show()

### Análisis 2: Todo el chat ###
with open(ruta_chat, 'r', encoding='utf-8') as archivo:
    chat = archivo.readlines()

# Contadores generales
contador_mensajes = Counter()
caracteres_totales = Counter()
mensajes_por_fecha = Counter()

for linea in chat:
    match = re.match(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] (\w+):", linea)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y").date()
        autor = match.group(2)
        contador_mensajes[autor] += 1
        caracteres_totales[autor] += len(linea)
        mensajes_por_fecha[fecha] += 1

# Mostrar estadísticas del análisis 2
print("\nEstadísticas generales del chat:")
print(f"Mensajes por persona: {contador_mensajes}")
print(f"Total de caracteres por persona: {caracteres_totales}")
print(f"Día con más mensajes: {max(mensajes_por_fecha, key=mensajes_por_fecha.get)} con {max(mensajes_por_fecha.values())} mensajes")

# Visualización: Mensajes por persona
plt.figure(figsize=(8, 6))
plt.bar(contador_mensajes.keys(), contador_mensajes.values())
plt.title("Mensajes por persona")
plt.xlabel("Persona")
plt.ylabel("Cantidad de mensajes")
plt.show()

# Visualización: Mensajes por fecha
fechas, mensajes = zip(*sorted(mensajes_por_fecha.items()))
plt.figure(figsize=(10, 6))
plt.plot(fechas, mensajes)
plt.title("Cantidad de mensajes por fecha")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de mensajes")
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
from collections import Counter
from datetime import datetime
import matplotlib.pyplot as plt

# Ruta del archivo
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"

# Leer el archivo
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Procesar datos
datos = []
for frase in frases:
    match = re.match(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] (\w+): (.+)", frase)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y")
        mes_año = fecha.strftime("%Y-%m")
        autor = match.group(2)
        texto = match.group(3).strip()
        datos.append({"Mes": mes_año, "Autor": autor, "Frase": texto})

# Crear un DataFrame
df = pd.DataFrame(datos)

# Tabla: Frecuencia de frases por frase y mes
frecuencia_frases = df.groupby(["Frase", "Mes"]).size().reset_index(name="Frecuencia")
frecuencia_total = df["Frase"].value_counts().reset_index()
frecuencia_total.columns = ["Frase", "Total"]

# Merge para agregar totales por frase
frecuencia_frases = frecuencia_frases.merge(frecuencia_total, on="Frase", how="left")

# Ordenar por las frases más repetidas
frecuencia_frases = frecuencia_frases.sort_values(by=["Total", "Frecuencia"], ascending=False)

# Mostrar las tablas
print("Frecuencia de frases por frase y mes:")
print(frecuencia_frases)

# Gráficos
# 1. Frases más repetidas (Top 10)
top_frases = frecuencia_total.head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_frases["Frase"], top_frases["Total"])
plt.title("Frases románticas más repetidas")
plt.xlabel("Frase")
plt.ylabel("Frecuencia Total")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Frecuencia mensual de las frases más repetidas
top_frases_mes = frecuencia_frases[frecuencia_frases["Frase"].isin(top_frases["Frase"])]
for frase in top_frases_mes["Frase"].unique():
    data = top_frases_mes[top_frases_mes["Frase"] == frase]
    plt.plot(data["Mes"], data["Frecuencia"], label=frase)

plt.figure(figsize=(12, 6))
plt.title("Frecuencia mensual de frases más repetidas")
plt.xlabel("Mes")
plt.ylabel("Frecuencia")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
from datetime import datetime

# Ruta del archivo
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"

# Leer el archivo
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Procesar datos
datos = []
for frase in frases:
    match = re.match(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] (\w+): (.+)", frase)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y")
        mes_año = fecha.strftime("%Y-%m")
        texto = match.group(3).strip().lower()
        datos.append({"Mes": mes_año, "Frase": texto})

# Crear un DataFrame
df = pd.DataFrame(datos)

# Filtrar frases relevantes
amor_frecuencia = df[df["Frase"].str.contains(r"\bamor\b", case=False)].groupby("Mes").size().reset_index(name="Frecuencia Amor")
te_amo_frecuencia = df[df["Frase"].str.contains(r"\bte amo\b", case=False)].groupby("Mes").size().reset_index(name="Frecuencia Te amo")

# Combinar las frecuencias en una tabla comparativa
comparativa = pd.merge(amor_frecuencia, te_amo_frecuencia, on="Mes", how="outer").fillna(0)
comparativa["Frecuencia Amor"] = comparativa["Frecuencia Amor"].astype(int)
comparativa["Frecuencia Te amo"] = comparativa["Frecuencia Te amo"].astype(int)

# Ordenar por mes
comparativa = comparativa.sort_values(by="Mes")

# Guardar la tabla como archivo CSV
ruta_salida = r"C:\Users\user\Desktop\ñucita_cabezona\Comparativa_Amor_vs_Te_Amo.csv"
comparativa.to_csv(ruta_salida, index=False, encoding='utf-8')
print(f"Tabla comparativa guardada en: {ruta_salida}")
```

```python
import re
import pandas as pd
from datetime import datetime
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk import tokenize
import matplotlib.pyplot as plt

# Paso 1: Cargar el archivo
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Paso 2: Extraer información de fecha y texto
data = []
for frase in frases:
    match = re.match(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] (\w+): (.+)", frase)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y")
        mes_anio = fecha.strftime("%Y-%m")
        texto = match.group(3).strip()
        data.append({"Mes": mes_anio, "Texto": texto})

df = pd.DataFrame(data)

# Paso 3: Análisis de sentimiento con NLTK (VADER)
sia = SentimentIntensityAnalyzer()

# Función para obtener puntaje de sentimiento
def obtener_sentimiento(texto):
    scores = sia.polarity_scores(texto)
    return scores["compound"]  # Puntaje neto de -1 a 1

df["Sentimiento"] = df["Texto"].apply(obtener_sentimiento)

# Paso 4: Cálculo de promedio y desviación estándar por mes
estadisticas_mensuales = df.groupby("Mes")["Sentimiento"].agg(["mean", "std"]).reset_index()

# Paso 5: Visualización de tendencia de sentimiento
plt.figure(figsize=(10, 6))
plt.plot(estadisticas_mensuales["Mes"], estadisticas_mensuales["mean"], marker='o', label="Promedio de sentimiento")
plt.fill_between(
    estadisticas_mensuales["Mes"],
    estadisticas_mensuales["mean"] - estadisticas_mensuales["std"],
    estadisticas_mensuales["mean"] + estadisticas_mensuales["std"],
    alpha=0.2,
    label="Desviación estándar"
)
plt.title("Tendencia de sentimiento por mes")
plt.xlabel("Mes")
plt.ylabel("Sentimiento promedio")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Palabras clave románticas
palabras_romanticas = [
    "amor", "cariño", "te amo", "te quiero", "mi vida", "mi cielo",
    "mi amor", "te extraño", "corazón", "beso", "abrazar", "linda",
    "preciosa", "hermosa", "ñu", "añuñin", "gordita", "te adoro", 
    "princesa", "mi reina", "mi rey", "mi bebé", "mi niña", 
    "mi niño", "te necesito", "mi todo", "eres especial"
]

# Lectura del archivo
ruta_frases = r"C:\Users\user\Desktop\ñucita_cabezona\Frases_Romanticas.txt"
with open(ruta_frases, 'r', encoding='utf-8') as archivo:
    frases = archivo.readlines()

# Estructura de datos
data = []
for frase in frases:
    # Extracción de fecha y texto
    match = re.match(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] (\w+): (.+)", frase)
    if match:
        fecha = datetime.strptime(match.group(1), "%d/%m/%y")
        mes_anio = fecha.strftime("%Y-%m")
        texto = match.group(3).strip().lower()

        # Conteo de palabras
        palabras = re.findall(r"\w+", texto)
        total_palabras = len(palabras)
        # Conteo de palabras románticas
        conteo_romanticas = sum(1 for p in palabras if p in palabras_romanticas)

        data.append({
            "Mes": mes_anio,
            "TotalPalabras": total_palabras,
            "Romanticas": conteo_romanticas
        })

df = pd.DataFrame(data)

# Suma por mes
agg_df = df.groupby("Mes").sum().reset_index()

# Cálculo de proporción
agg_df["ProporcionRomanticas"] = agg_df["Romanticas"] / agg_df["TotalPalabras"]

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(agg_df["Mes"], agg_df["ProporcionRomanticas"], marker='o')
plt.title("Proporción de palabras románticas por mes")
plt.xlabel("Mes")
plt.ylabel("Proporción romántica")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
from datetime import datetime

# Ruta del archivo de chat
ruta_archivo = r"C:\Users\user\Desktop\ñucita_cabezona\_chat.txt"

# Lectura de líneas
with open(ruta_archivo, 'r', encoding='utf-8') as f:
    lineas = f.readlines()

reg_patron = re.compile(r"\[(\d{2}/\d{2}/\d{2}) \d{2}:\d{2}:\d{2}\] .*?: .*")

fechas = []
for linea in lineas:
    match = reg_patron.match(linea)
    if match:
        fecha_str = match.group(1)
        # Convertir la fecha al formato datetime
        # Ajusta la máscara si tu archivo usa otro orden (día/mes/año vs. mes/día/año).
        fecha_dt = datetime.strptime(fecha_str, "%d/%m/%y")
        fechas.append(fecha_dt)

# Crear DataFrame con las fechas
df = pd.DataFrame({"Fecha": fechas})
df["Año"] = df["Fecha"].dt.year
df["Mes"] = df["Fecha"].dt.month
df["Día"] = df["Fecha"].dt.day

# Contar mensajes por año-mes-día
conteo_diario = df.groupby(["Año", "Mes", "Día"]).size().reset_index(name="TotalMensajes")

# Filtrar solo meses de junio (6) a diciembre (12)
conteo_filtrado = conteo_diario[conteo_diario["Mes"].between(6, 12)]

# Para cada año-mes, hallar el día con más mensajes
resultado = []
for (anio, mes), grupo_mes in conteo_filtrado.groupby(["Año", "Mes"]):
    # Seleccionar la fila con mayor número de mensajes
    fila_max = grupo_mes.loc[grupo_mes["TotalMensajes"].idxmax()]
    resultado.append({
        "Año": anio,
        "Mes": mes,
        "Día": fila_max["Día"],
        "TotalMensajes": fila_max["TotalMensajes"]
    })

# Convertir a DataFrame y mostrar
df_resultado = pd.DataFrame(resultado)
df_resultado = df_resultado.sort_values(by=["Año", "Mes"], ascending=True)

print("Día con más mensajes en cada mes (junio a diciembre):")
print(df_resultado.to_string(index=False))
```
